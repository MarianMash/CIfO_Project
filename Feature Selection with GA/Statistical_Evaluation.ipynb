```python

import pandas as pd
import glob
import plotly.graph_objects as go
import numpy as np
import re

path = r'/Users/masha/Desktop/NOVA/Optimization/Project_CIfO/new_csv' # my path
all_files = glob.glob(path + "/*.csv")

di = dict()

for filename in all_files:
    one_one = pd.read_csv(filename)
    fig = go.Figure()
    for col in one_one.columns[1:]:

        name = re.split("/|\.",filename)[-2]     
        fig.add_trace(go.Scatter(x=one_one.index, y=one_one[col],
                        mode='lines',
                        name=col))
    fig.update_layout(
    title=name,
    xaxis_title="Generation",
    yaxis_title="Fitness")




    fig.show()
    #identify mean of the best fitness of the last generation for this combination
    the_mean = np.mean(one_one.iloc[-1].to_list()[1:])
    median_of_df = np.median(one_one.iloc[-1].to_list()[1:])


    a_list = list()
    b_list = list()
    for col in one_one.columns[1:]:
        a_list.append(one_one[col].idxmax())

    first_appearance = np.mean(a_list)
    print(f"The mean of this combination is: {round(the_mean,3)} and the mean of first appearance {first_appearance}")
    di[name] = [round(the_mean,3),round(median_of_df,3),first_appearance,]
    



```

```python
new_df = pd.DataFrame.from_dict(di,orient='index',columns = ["AVG_fitness","Median","AVG_firstMAX"])
new_df.sort_values("AVG_fitness", ascending = False)
```
